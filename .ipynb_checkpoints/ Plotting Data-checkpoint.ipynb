```python
import numpy as np
import matplotlib.pyplot as plt
```

#  Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)  # Plotting the line graph
plt.show()  # Displaying the graph
```

```python
plt.plot(x, y, marker='o')  # Plotting the line graph with markers
plt.show()  # Displaying the graph
```

```python
plt.plot(x, y, marker='o')  # Plotting the line graph with markers

# Adding title and labels
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()  # Displaying the graph
```

# Styling Plots

```python
plt.plot(x, y, marker='o', linestyle='--', color='red')  # Plotting with dashed line and red color

# Adding title and labels
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()  # Displaying the graph
```

```python
plt.plot(x, y, marker='s', linestyle='--', color='red', markersize=10, markerfacecolor='green') # Adjusting marker size to 10 & color to Green

# Adding title and labels
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
# Plotting with star markers and blue color
plt.plot(x, y, marker='*', linestyle='--', color='red', markersize=10, markerfacecolor='blue')

# Adding title and labels
plt.title('Example Graph with Star Markers')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()

# Plotting with diamond markers and black color
plt.plot(x, y, marker='D', linestyle='--', color='red', markersize=10, markerfacecolor='black')

# Adding title and labels
plt.title('Example Graph with Diamond Markers')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

# Creating the bar chart
plt.bar(products, sales)

# Adding labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of Different Products")

# Showing the plot
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

# Creating the bar chart with customizations
plt.bar(products, sales, color='skyblue', width=0.5)

# Adding labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of Different Products")

# Showing the plot
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

# Creating the scatter plot
plt.scatter(x, y)

# Adding labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Random Data")

# Showing the plot
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100  # Varying marker sizes

plt.scatter(x, y, s=sizes, c=y, cmap='viridis')  # Colormap based on y-values
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Scatter Plot")
plt.colorbar()  # Show color scale
plt.show()
```

# Subplots

```python
# Creating the figure and subplots
fig, axes = plt.subplots(2, 1)

# First subplot - Line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0].plot(x, y)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# Second subplot - Histogram
data = np.random.randn(1000)
axes[1].hist(data)
axes[1].set_title('Histogram of Random Data')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

# Adjusting layout
plt.tight_layout()
plt.show()
```
